## Wavelets para procesamiento de señales sEMG

Recordemos para que sirve las wavelets?

Al usar la transformada de Fourier funcionará muy bien cuando el espectro de frecuencias esté estacionario. Es decir, las frecuencias presentes en la señal no dependen del tiempo.Pero la mayoría de las señales que se ve en la vida real son de naturaleza no estacionaria. Como ECG, sensores, etc. La cuestion con la transformada de Fourier es que tiene una alta resolución en el dominio de la frecuencia, pero una resolución nula en el dominio del tiempo. Esto significa que puede decirnos exactamente qué frecuencias están presentes en una señal, pero no en qué lugar en el tiempo se han producido estas frecuencias.

Un mejor enfoque para analizar señales con un espectro de frecuencia dinámico es la Transformada Wavelet. No solo nos dice qué frecuencias están presentes en una señal, sino también en qué momento se han producido estas frecuencias. Esto se logra trabajando con diferentes escalas. Primero se mira la señal con una escala y se analiza las características "grandes" y luego se mira la señal con escalas más pequeñas para analizar las características más pequeñas.

<img src="wavelet.gif" width="650" align="center">

En otras palabras, elegimos una wavelet de una escala particular (como la wavelet azul en el gif de arriba). Luego, deslizamos esta ondícula a través de toda la señal, es decir, variamos su ubicación, donde en cada paso de tiempo multiplicamos la ondícula y la señal. El producto de esta multiplicación nos da un coeficiente para esa escala de ondículas en ese paso de tiempo. Luego aumentamos la escala de wavelets (por ejemplo, las wavelets roja y verde) y repetimos el proceso.



### Eliminación de ruido y extracción de caracteristicas

Las señales EMG requieren un método confiable y preciso para cada uno de los siguientes 4 pasos:



La transformada wavelets es una de las herramientas de procesamiento de señales más poderosas. Para el reconocimiento de patones EMG, permite la extracción de características a partir de la descomposición con wavelets de múltiples niveles de la señal EMG. 

Se utilizan diferentes niveles de varias wavelets madre para obtener los componentes de resolución útiles de la señal EMG. Se selecciona el componente de resolución EMG óptima (sub-señal) y luego se realiza la reconstrucción de la señal de información útil. Eliminando asi, el ruido y las piezas de EMG no deseadas  

De la señal EMG estimada que es una parte EMG efectiva se extraen las características populares, es decir, el valor absoluto medio y la raíz cuadrada media, con el fin de mejorar la calidad de la separabilidad de clases. 



Utilizar wavelets en aplicaciones de EMG permite mejorar la separabilidad de clases en el espacio de características y garantiza que el resultado de la precisión de la clasificación de patrones sea lo más alto posible.



### Transformación wavelet y métodos de extracción de características

El método de transformación wavelet se divide en dos tipos: transformada de ondícula discreta (DWT) y transformada de ondícula continua (CWT). 

La DWT es una técnica que transforma iterativamente una señal interesada en subconjuntos de coeficientes de resolución múltiple. El DWT transforma la señal EMG con una función de base de ondículas (WF) adecuada. Por lo tanto, el análisis parte de la resolución múltiple mediante el estudio de las características EMG con diferentes escalas y variaciones locales y también la eliminación de los componentes de frecuencia no deseados.

Procedimiento:

- La señal EMG original (S) pasa a través de un filtro de paso bajo y un filtro de paso alto (los coeficientes de los filtros dependen del tipo de WF) 
- Se obtiene un subconjunto de coeficientes de aproximación (cA1) y un subconjunto de coeficientes de detalle (cD1) en el primer nivel. Este proceso se repite hasta obtener el nivel final deseado. 
- Cuatro niveles de descomposición de wavelets muestran un mejor rendimiento 
- Esto genera los subconjuntos de coeficientes de la aproximación de nivel 4 (cA4) y los detalles de nivel 1, 2, 3, 4 (cD1, cD2, cD3 y cD4), respectivamente. Además, cada subconjunto de coeficientes se puede reconstruir para obtener una parte de señal EMG efectiva. 
- La reconstrucción de una señal se realiza utilizando la transformada wavelet inversa utilizando los coeficientes de todos los componentes de la descomposición de nivel final, es decir, la aproximación de cuarto nivel y los primeros cuatro niveles de detalle (cA4, cD1, cD2, cD3 y cD4).

### Eliminacion de ruido:

La eliminación de ruido clásica de wavelets se basa en la asignación de un valor de umbral que se puede dividir en el umbral suave y el umbral duro. La eliminación de ruido se puede dividir en los siguientes pasos:

• Realizar una descomposición de múltiples escalas de las señales EMG sin procesar, para observar los coeficientes de ondas de la señal
• Estimar el ruido y elegir el umbral, realizar el análisis de umbral a los coeficientes de wavelet y obtener nuevos coeficientes
• Reconstruir las señales EMG por el coeficiente de wavelet revisado




### Extraccion de caracteristicas y clasificacion de patrones:

Existen una variedad de clasificadores, de los cuales se pueden destacar:

- Clasificador  Bayesiano. Es el método  estadístico clásico. Donde se estima la probabilidad a posteriori de la pertenencia de una muestra de prueba a una de las clases dadas. Esta probabilidad  es evaluada en función  de las probabilidades a priori de cada clase y la probabilidad  condicional, resultante de la distribución de las muestras de entrenamiento en cada clase y evaluada según  el teorema de Bayes.



- Análisis  discriminante  lineal (LDA Lineal discriminant analysis). Es otro método estadístico que discrimina la separabilidad  de clases en forma lineal, lo cual no se ajusta de manera eficiente a los patrones EMGS por su dinámica estocástica.



- Redes neuronales artificiales (ANN Artificial Neural Network). Las redes ANN tienen  características como: generalización y habilidad  para aprender de la experiencia sin  requerir un modelo  matemático a priori de las características de la señal, adaptación a los cambios de condiciones ambientales y habilidad  para procesar datos degradados o  incompletos, lo que las hace apropiadas para señales críticas como las señales EMGS. Sin  embargo, las redes ANN requieren de un gran número de muestras y mucho tiempo de entrenamiento para un aprendizaje adecuado.



- Lógica  difusa. Son sistemas de inferencia, y al igual que las ANN simulan el comportamiento de decisión humana. Con posibilidad  de introducir la experiencia de un experto en la elaboración de reglas de decisión, en un esquema IF_THEN. Estos sistemas se pueden clasificar en dos categorías: 

    1. Modelos lingüísticos basado en colección de reglas IF_THEN, cuyos antecedentes y consecuentes (modelo Mamdani)
    2. Sistemas tipo Tagagi Sugeno donde usa una estructura de reglas que tienen antecedentes fuzzy y consecuentes funcionales. 
    



- Sistemas  Neuro fuzzy. La combinación de redes neuronales y lógica difusa permite una relación simbiótica en la cual se aprovecha el conocimiento de un experto y la capacidad de aprendizaje y eficiencia computacional de la red neuronal, logrando un sistema de decisiones más inteligentes.

### Ejemplo 2: extrayendo frecuencias altas de una señal

```python
#Fuente: https://github.com/EstebanHess/WaveletTransform_with_pywt

import math
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import pylab
import pywt
```

```python
f_1 = 4
w = 2. * np.pi * f_1
time_interval = 1
fig = pylab.figure()
signals_lowfreq = {}
for i, samples in enumerate((5, 14, 200, 5000)):
    pylab.subplot(2, 2, i+1)
    pylab.title('%i samples'%samples)
    t = np.linspace(0, time_interval, samples)
    y = np.sin(w * t)
    data_points = str(samples) + '_points'
    signals_lowfreq[data_points] = y
    pylab.plot(t, y, '.-')
    plt.subplots_adjust(bottom=0.4, right=2, top=2)
fig.show()
```

```python
f_2 = f_1*4
w = 2. * np.pi * f_2
time_interval = 1
fig = pylab.figure()
signals_highfreq = {}
for i, samples in enumerate((5, 14, 200, 5000)):
    pylab.subplot(2, 2, i+1)
    pylab.title('%i samples'%samples)
    t = np.linspace(0, time_interval, samples)
    y = np.sin(w * t)
    data_points = str(samples) + '_points'
    signals_highfreq[data_points] = y
    pylab.plot(t, y, '.-')
    plt.subplots_adjust(bottom=0.4, right=2, top=2)
fig.show()
```

```python
sine_wave_low = signals_lowfreq['200_points']
sine_wave_high = signals_highfreq['200_points']
```

### Transformada wavelet discreta:

Descomponer una señal en componentes que contienen diferentes frecuencias: crear una señal agregando las dos señales anteriores: frecuencia baja constante más una señal mixta (constante en la primera parte y alta frecuencia en la segunda parte)

```python
zero_streak = np.zeros(100)
zero_high = np.concatenate((zero_streak,signals_highfreq['200_points'][100:]))

sine_wave_low = signals_lowfreq['200_points']
sine_wave_high = signals_highfreq['200_points']

pylab.subplot(2,2,1)
pylab.plot(sine_wave_low)
pylab.subplot(2,2,2)
pylab.plot(zero_high)
fig.show()
```

```python
signal_mixed_half = signals_lowfreq['200_points'] + zero_high
plt.plot(signal_mixed_half)
```

Descomponer la señal mixta (signal_mixed_half) para volver a extraer la señal de baja frecuencia original (sine_wave_low):
Visualización de la composición de la frecuencia utilizando la transformada de ondículas discretas

Fuente: https://github.com/PyWavelets/pywt/blob/master/demo/dwt_signal_decomposition.py

```python
mode = pywt.Modes.smooth

def plot_signal_decomp(data, w, title):
    """Decompose and plot a signal S.
    S = An + Dn + Dn-1 + ... + D1
    """
    w = pywt.Wavelet(w)
    a = data
    ca = []
    cd = []
    for i in range(10):
        (a, d) = pywt.dwt(a, w, mode)
        ca.append(a)
        cd.append(d)

    rec_a = []
    rec_d = []

    for i, coeff in enumerate(ca):
        coeff_list = [coeff, None] + [None] * i
        rec_a.append(pywt.waverec(coeff_list, w))

    for i, coeff in enumerate(cd):
        coeff_list = [None, coeff] + [None] * i
        rec_d.append(pywt.waverec(coeff_list, w))

    fig = plt.figure(figsize=(20,10))
    ax_main = fig.add_subplot(len(rec_a) + 1, 1, 1)
    ax_main.set_title(title)
    ax_main.plot(data)
    ax_main.set_xlim(0, len(data) - 1)

    for i, y in enumerate(rec_a):
        ax = fig.add_subplot(len(rec_a) + 1, 2, 3 + i * 2)
        ax.plot(y, 'r')
        ax.set_xlim(0, len(y) - 1)
        ax.set_ylabel("A%d" % (i + 1))

    for i, y in enumerate(rec_d):
        ax = fig.add_subplot(len(rec_d) + 1, 2, 4 + i * 2)
        ax.plot(y, 'g')
        ax.set_xlim(0, len(y) - 1)
        ax.set_ylabel("D%d" % (i + 1))
```

```python
plot_signal_decomp(signal_mixed_half, 'sym14', "Signal decomposition")
```

El siguiente código ejecuta la misma operación que se visualiza arriba.

```python
def dwt_steps(a, w, mode, num_steps):
    ca = []
    cd = []
    for i in range(num_steps + 1):
        (a, d) = pywt.dwt(a, w, mode)
        ca.append(a)
        cd.append(d)
    return ca, cd
```

```python
w = 'sym14'
mode = pywt.Modes.smooth
ca, cd = dwt_steps(signal_mixed_half, w, mode, 10)
```

Reconstruyendo la señal a partir de las estimaciones:

```python
rec_a = []
rec_d = []

for i, coeff in enumerate(ca):
    coeff_list = [coeff, None] + [None] * i
    rec_a.append(pywt.waverec(coeff_list, w))

for i, coeff in enumerate(cd):
    coeff_list = [None, coeff] + [None] * i
    rec_d.append(pywt.waverec(coeff_list, w))
    
low_freq_rec = rec_a[2][:200]
high_freq_rec = signal_mixed_half - low_freq_rec
```

Los dos componentes de la señal reconstruidos por la transformada wavelet son bastante similares a los dos componentes originales de la señal que se utilizaron para crear la señal mixta:

```python
pylab.subplot(2,2,1)
pylab.plot(low_freq_rec)
pylab.subplot(2,2,2)
pylab.plot(high_freq_rec)
fig.show()
```
