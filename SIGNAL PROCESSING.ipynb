## Extracción de caracteristicas y clasificadores

La Transformada de Fourier permite transformar una señal de su dominio de tiempo a su dominio de frecuencia. Los picos en el espectro de frecuencias indican las frecuencias más frecuentes en la señal. Cuanto más grande y agudo es un pico, más prevalente es una frecuencia en una señal. La ubicación (valor de frecuencia) y la altura (amplitud) de los picos en el espectro de frecuencia se pueden utilizar como entrada para clasificadores

#### Wavelet packet transform (Wavelet packet decomposition)

Al usar la transformada de Fourier funcionará muy bien cuando el espectro de frecuencias esté estacionario. Es decir, las frecuencias presentes en la señal no dependen del tiempo.Pero la mayoría de las señales que se ve en la vida real son de naturaleza no estacionaria. Como ECG, sensores, etc. La cuestion con la transformada de Fourier es que tiene una alta resolución en el dominio de la frecuencia, pero una resolución nula en el dominio del tiempo. Esto significa que puede decirnos exactamente qué frecuencias están presentes en una señal, pero no en qué lugar en el tiempo se han producido estas frecuencias.

Un mejor enfoque para analizar señales con un espectro de frecuencia dinámico es la Transformada Wavelet. No solo nos dice qué frecuencias están presentes en una señal, sino también en qué momento se han producido estas frecuencias. Esto se logra trabajando con diferentes escalas. Primero se mira la señal con una escala y se analiza las características "grandes" y luego se mira la señal con escalas más pequeñas para analizar las características más pequeñas.



La Transformada Wavelet utiliza una serie de funciones llamadas wavelets, cada una con una escala diferente. Wavelet significa onda pequeña.
Dado que el Wavelet está localizado en el tiempo, se puede multiplicar la señal con el wavelet en diferentes ubicaciones en el tiempo (convolucion).

# Independent Component Analysis

Análisis de componentes independientes (ICA) - analogía:



ICA es un método computacional para separar una señal multivariante en sus componentes subyacentes. Con ICA, podemos extraer el componente deseado de la combinación de múltiples señales.

### Algoritmo de análisis de componentes independientes (ICA)

En un nivel alto, ICA se puede dividir en los siguientes pasos.

1. Centro x restando la media
2. Blanquear x
3. Elija un valor inicial aleatorio para la matriz de desmezcla w
4. Calcule el nuevo valor de w
5. Normalizar w
6. Compruebe si el algoritmo ha convergido y, si no lo ha hecho, vuelva al paso 4
7. Tome el producto escalar de w y x para obtener las señales de fuente independientes



### Blanqueo

Antes de aplicar el algoritmo ICA, primero debemos " blanquear" nuestra señal. A “blanquear” entenderá la señal emitida que lo transformamos en una forma tal que las correlaciones potenciales entre sus componentes se eliminarán (covarianza igual a 0) y la varianza de cada componente es igual a 1. Otra forma de verlo es que la La matriz de covarianza de la señal blanqueada será igual a la matriz de identidad.



Esto implica la descomposición del valor propio de su matriz de covarianza. La ecuación matemática correspondiente se puede describir como sigue.



donde D es una matriz diagonal de valores propios (cada lambda es un valor propio de la matriz de covarianza)



y E es una matriz ortogonal de vectores propios



Una vez se preprocesa la señal, para cada componente, se actualizan los valores de la matriz de desmezcla w hasta que el algoritmo haya convergido o se haya alcanzado el número máximo de iteraciones. La convergencia se considera lograda cuando el producto escalar de **wy** su transposición es aproximadamente igual a 1.

```python
import numpy as np
np.random.seed(0)
from scipy import signal
from scipy.io import wavfile
from matplotlib import pyplot as plt
import seaborn as sns
sns.set(rc={'figure.figsize':(11.7,8.27)})
```

A continuación, definimos g y g’los cuales usaremos para determinar el nuevo valor de w.

```python
def g(x):
    return np.tanh(x)
def g_der(x):
    return 1 - g(x) * g(x)
```

Creamos una función para centrar la señal restando la media.

```python
def center(x):
    x = np.array(x)
    
    mean = x.mean(axis=1, keepdims=True)
    
    return x - mean
```

Definimos una función para blanquear la señal utilizando el método descrito anteriormente.

```python
def whitening(x):
    cov = np.cov(x)

    d, E = np.linalg.eigh(cov)

    D = np.diag(d)

    D_inv = np.sqrt(np.linalg.inv(D))

    x_whiten = np.dot(E, np.dot(D_inv, np.dot(E.T, x)))

    return x_whiten
```

Definimos una función para actualizar la matriz de desmezcla w.

```python
def calculate_new_w(w, X):
    w_new = (X * g(np.dot(w.T, X))).mean(axis=1) - g_der(np.dot(w.T, X)).mean() * w

    w_new /= np.sqrt((w_new ** 2).sum())

    return w_new
```

Finalmente, definimos el método principal que llama a las funciones de preprocesamiento, inicializa w a algún conjunto aleatorio de valores y actualiza iterativamente w . Nuevamente, la convergencia se puede juzgar por el hecho de que un w ideal sería ortogonal y, por lo tanto, w multiplicado por su transposición sería aproximadamente igual a 1. Después de calcular el valor óptimo de w para cada componente, tomamos el producto escalar del resultado matrix y la señal x para obtener las fuentes.

```python
def ica(X, iterations, tolerance=1e-5):
    X = center(X)
    
    X = whitening(X)
        
    components_nr = X.shape[0]

    W = np.zeros((components_nr, components_nr), dtype=X.dtype)

    for i in range(components_nr):
        
        w = np.random.rand(components_nr)
        
        for j in range(iterations):
            
            w_new = calculate_new_w(w, X)
            
            if i >= 1:
                w_new -= np.dot(np.dot(w_new, W[:i].T), W[:i])
            
            distance = np.abs(np.abs((w * w_new).sum()) - 1)
            
            w = w_new
            
            if distance < tolerance:
                break
                
        W[i, :] = w
        
    S = np.dot(W, X)
    
    return S
```

Definimos una función para graficar y comparar las señales originales, mixtas y predichas.

```python
def plot_mixture_sources_predictions(X, original_sources, S):
    fig = plt.figure()

    plt.subplot(3, 1, 1)
    for x in X:
        plt.plot(x)
    plt.title("mixtures")

    plt.subplot(3, 1, 2)
    for s in original_sources:
        plt.plot(s)
    plt.title("real sources")

    plt.subplot(3,1,3)
    for s in S:
        plt.plot(s)
    plt.title("predicted sources")
    
    fig.tight_layout()
    plt.show()
```

Creamos un método para mezclar artificialmente diferentes señales de fuente

```python
def mix_sources(mixtures, apply_noise=False):
    for i in range(len(mixtures)):
        
        max_val = np.max(mixtures[i])
        
        if max_val > 1 or np.min(mixtures[i]) < 1:
            
            mixtures[i] = mixtures[i] / (max_val / 2) - 0.5
            
    X = np.c_[[mix for mix in mixtures]]
    
    if apply_noise:
        
        X += 0.02 * np.random.normal(size=X.shape)
        
    return X
```

Luego, creamos 3 señales, cada una con su propio patrón distintivo.

```python
n_samples = 2000
time = np.linspace(0, 8, n_samples)
s1 = np.sin(2 * time)
s2 = np.sign(np.sin(3 * time)) # square signal
s3 = signal.sawtooth(2 * np.pi * time)
```

En el ejemplo anterior, calculamos el producto escalar de la matriz A y las señales para obtener una combinación de las tres. Luego usamos Análisis de Componentes Independientes para separar la señal mezclada en las señales de la fuente original.

```python
X = np.c_[s1, s2, s3]
A = np.array(([[1, 1, 1], [0.5, 2, 1.0], [1.5, 1.0, 2.0]]))
X = np.dot(X, A.T)
X = X.T
S = ica(X, iterations=1000)
actual = mix_sources([s1,s2,s3])
plot_mixture_sources_predictions(X, [s1, s2, s3], S)
```

```python
from sklearn.decomposition import FastICA

np.random.seed(0)
n_samples = 2000
time = np.linspace(0, 8, n_samples)

s1 = np.sin(2 * time)
s2 = np.sign(np.sin(3 * time))
s3 = signal.sawtooth(2 * np.pi * time)

S = np.c_[s1, s2, s3]
S += 0.2 * np.random.normal(size=S.shape)
S /= S.std(axis=0)
A = np.array([[1, 1, 1], [0.5, 2, 1.0], [1.5, 1.0, 2.0]])
X = np.dot(S, A.T)

ica = FastICA(n_components=3)
S_ = ica.fit_transform(X)

fig = plt.figure()

models = [X, S, S_]

names = ['mixtures', 'real sources', 'predicted sources']

colors = ['red', 'blue', 'orange']

for i, (name, model) in enumerate(zip(names, models)):
    plt.subplot(4, 1, i+1)
    plt.title(name)
    for sig, color in zip (model.T, colors):
        plt.plot(sig, color=color)
        
fig.tight_layout()        
plt.show()
```
